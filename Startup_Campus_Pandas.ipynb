## 1 Creating Data

### 1.1 DataFrame
A DataFrame adalah sebuah tabel

```python
pd.DataFrame({'Yes': [50, 21], 'No': [131, 2]})
```

Index dalam dataframe dapat diubah sesuai kebutuhan

```python
pd.DataFrame({'Gender': ['laki-laki', 'laki-laki'],
              'Umur': [36, 27]},
             index=['Mas Ainun', 'Mas Zul'])
```

### 1.2 Series
A Series, by contrast, is a sequence of data values. If a DataFrame is a table, a Series is a list. And in fact you can create one with nothing more than a list:



```python
pd.Series([1, 2, 3, 4, 5])
```

A Series is, in essence, a single column of a DataFrame. So you can assign row labels to the Series the same way as before, using an `index` parameter. However, a Series does not have a column name, it only has one overall `name`:

```python
pd.Series([30, 35, 40], index=['2015 Sales', '2016 Sales', '2017 Sales'], name='Product A')
```

## 9 Grouping

```python
df_group = df[['State','Category','Sales','Quantity']]
```

```python
df_group.groupby('State')['Sales'].sum()
```

```python
df_group.groupby(['State','Category'])[['Sales','Quantity']].sum()
```

```python
df_group.groupby(['State','Category']).agg({'Sales':'sum','Quantity':'mean'})
```

## 10 Manipulating row and columns

### 10.1 Pivot

```python
pd.pivot_table(df,index='State', columns='Category', values='Sales')
```

### 10.2 Unstack

```python
df_group = df_group.groupby(['State','Category']).agg({'Sales':'sum','Quantity':'mean'})
df_group
```

```python
df_unstack = df_group.unstack(level=1)
df_unstack
```

### 10.3 Stack

```python
# Stack the prescribed level(s) from columns to index. (menumpuk)
df_stack = df_unstack.stack(level=1)
```
